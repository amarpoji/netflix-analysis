```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
netflix_filepath = ('netflix_titles.csv')

netflix_data = pd.read_csv(netflix_filepath)

netflix_data.head()
# netflix_data.info()
```

```python
netflix_data['listed_in'].value_counts().head(10)
```

```python
top_genres = netflix_data['listed_in'].value_counts().head(10)

top_genres.plot( kind= 'bar',figsize=(10,5), color='skyblue')




plt.title('top 10 most populor genre in netflix')
plt.xlabel('Genre')
plt.ylabel('Number of titles')
plt.xticks(rotation=90)
plt.show()


top_genres
```
